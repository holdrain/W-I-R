```python
import matplotlib.pyplot as plt
import numpy as np

# 


labels = np.array(['A', 'R', 'I'])  #1-Identity Leakage
model1 = np.array([0.9998,	0.7067,	0.1930*5]) # clean
# model2 = np.array([0.9999,	0.5672,	0.0606]) # Robust Training
model2 = np.array([0.9988,	1.0000,	0.1439*5]) 

model3 = np.array([0.9995,	0.99,	0.1907*5]) # Certified Watermarking

# 
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]  # 

# 
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(18, 6), subplot_kw=dict(polar=True))
plt.rc('font', size=28,weight='bold')  # 

# 
colors = [(216/255, 222/255, 183/255), (227/255, 215/255, 230/255),(179/255,203/255,220/255)]
# colors = ['#D175A9', '#A09CBF','#159D74']
# 
# for ax, model, title in zip(axs, [model1, model2, model3], ['Clean Vanilla', 'Robust Training', 'Certified Watermarking']):
for ax, model, color, title in zip(axs, [model1, model2, model3], colors, ['Clean Model', 'Robust Training', 'W-IR']):
    model = np.concatenate((model, [model[0]]))  # 
    # ax.fill(angles, model, alpha=0.25)
    ax.fill(angles, model, color=color, alpha=1.0)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels,fontsize=28)
    ax.spines['polar'].set_visible(True)
    ax.spines['polar'].set_linewidth(3)

    ax.set_yticks([0.50, 0.60,0.70,0.80,0.90,1.00])  # 
    ax.set_yticklabels(['','','','','',''],fontsize=24)
    ax.set_ylim(0.5, 1)
    ax.set_title(title,fontsize=24,pad=20,fontweight='bold')

# full_labels = ['E-Effectiveness', 'R-Robustness', 'I-Identity Protetion']
# plt.legend(full_labels, loc='lower right', bbox_to_anchor=(1.3, -0))
# plt.subplots_adjust(left=0.1, right=0.9, top=0.75, bottom=0.1) 
plt.tight_layout()  # 
plt.savefig('intro.pdf', format='pdf', dpi=300)  # 300 DPI
plt.show()
```
